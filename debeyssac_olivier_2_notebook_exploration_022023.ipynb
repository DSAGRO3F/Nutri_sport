# Domaine de l'alimentation - athlètes de haut niveau.


## Introduction.
- Lorsque les apports alimentaires ne peuvent pas être couverts de manière appropriée (par exemple, si l’athlète est en voyage dans un pays où l’offre alimentaire est limitée) ou lorsqu’un sportif souffre d’un déficit particulier, vitaminique ou minéral, des compléments et/ou des suppléments peuvent être alors recommandés.
- Les suppléments peuvent être classés en quatre groupes, du groupe A (autorisés et dont les preuves scientifiques sont avérées) au groupe D (interdits par l’Agence mondiale antidopage, AMA).
- De nombreux minéraux, vitamines et acides aminés sont quotidiennement présents dans notre assiette et il est préférable, en tout premier lieu, d’équilibrer son alimentation avant de se complémenter.
- Les durées quotidiennes d’entraînement, souvent supérieures à 4 heures, associées à de fortes intensités d’exercices, éprouvent l’organisme. Des apports adéquats en énergie, mais aussi en fer, cuivre, manganèse, magnésium, sélénium, zinc, et en vitamines A, B6, B12, C et E, sont donc particulièrement importants pour la santé et la performance sportive.
- Ces éléments doivent être fournis par une alimentation variée, fondée pour une large part sur des nutriments riches, tels que les légumes, les fruits, les fèves, les graines, les viandes maigres, les poissons, les produits laitiers et les huiles non saturées.
- Les études scientifiques montrent que, pour une grande majorité d’entre eux (plus de 70 %), les athlètes atteignent les apports recommandés en vitamines et minéraux en mangeant régulièrement des aliments variés.
- Classification des aliments et nutriments en fonction de leurs effets prouvés et non prouvés sur la performance sportive:


- Groupe A : suppléments et aliments du sport avec preuves scientifiques: Les constituants de ce groupe sont une source d’énergie et de nutriments utiles et opportuns dans l’alimentation du sportif, ou ont montré des preuves scientifiques de leurs effets positifs sur la performance.
    - bicarbonate et citrate,
    - caféine,
    - calcium,
    - magnésium,
    - créatine,
    - fer,
    - électrolytes à remplacer (Cf. Fiche no 10 L’hydratation, p. 332),
    - multivitamines et minéraux,
    - barres et gels énergétiques,
    - boissons de l’effort (Cf. Fiche no 10 L’hydratation, p. 332),
    - protéines de petit-lait.

- Groupe B : suppléments et aliments du sport dont les preuves scientifiques quant à l’effet sur la performance sont controversées ou obtenues à partir de résultats préliminaires. Les constituants de ce groupe ont reçu une attention scientifique particulière et ont quelquefois fait l’objet d’expérimentations avec des populations d’athlètes. Les résultats préliminaires des études suggèrent des résultats bénéfiques pour la performance.
    - vitamines antioxydantes, C et E,
    - colostrum,
    - β-alanine,
    - glutamine,
    - mélatonine,
    - ribose,
    - probiotiques (Cf. Fiche no 8 L’immunité, p. 305),
    - quercétine,
    - jus de betterave,
    - huile de poisson.

Il existe deux autres groupes d'aliments. Mais leur influence positive sur la performance sportive n'est pas démontrée. Pour le quatrième groupe, ces aliments et nutriments sont interdits par l'agence anti-dopage internationale. On se concentrera sur les aliments et nutriments du groupe 1 pour lequel il existe des preuves scientifiques d'une influence positive sur la performance sportive, ainsi que le groupe 2.

```python
# Concaténation des sets df_num, df_cat.

def verif(df_1, df_2):
    print("dimensions set numérique:{}".format(df_1.shape))
    print("dimensions set catégoriel:{}".format(df_2.shape))

    df_conc = pd.concat([df_1, df_2], axis=1)
    
    # dimensions de df_conc:
    long_c = df_conc.shape[0]
    larg_c = df_conc.shape[1]
    
    # vérification dimensions:
    long = df_1.shape[0]
    larg = df_1.shape[1] + df_2.shape[1]
    
    if long != long_c:
        print("dimension axe 0 fausse")
    else:
        print("ok")
        
    if larg != larg_c:
        print("dimension axe 1 fausse")
        print("ok")
        
    return df_conc


```

```python
# Fonction détection outliers. Distribution non normale ==> donc pas utilisation z-score, mais IQR.
def outliers(df):
    l_outliers=[]
    df_num = df.select_dtypes('float64')
    df_cat = df.select_dtypes('object')
    df_cat = df_cat.loc[:, ['code', 'product_name', 'brands']]
    
    
    for i in range(len(df_num.columns)):
        col_name = df_num.columns[i]
        df = pd.concat([df_cat,df_num.loc[:,col_name]], axis=1)
        
        # calcul quartile 1 et 3, IQR
        q_1 = df_num.loc[:,col_name].quantile(0.25)
        q_3 = df_num.loc[:,col_name].quantile(0.75)
        
        IQR = q_3 - q_1
        limite = q_3 + 1.5 * IQR
        
        l_outliers.append(df[df_num[col_name]>limite])
                

    return l_outliers
    
```

```python
def calc_quantiles(df):
    
    nb_mesures = 4
    val_mesures = [0] * nb_mesures
    val_quantiles = [0] * nb_mesures
    l_df = []
    
    df = df.select_dtypes('float64')

    for i in range(len(df.columns)):
        col_name = df.columns[i]        
        quantile = col_name
        
        df_var = df[col_name]
        df_var = df_var.dropna()
        
        # Calcul des quantiles              
        df_q1 = df_var.quantile(0.25)
        df_q2 = df_var.quantile(0.50)
        df_q3 = df_var.quantile(0.75)
        df_q4 = df_var.quantile(1)
        
        # Création listes pour contenir les variables des valeurs remarquables.        
        val_quantiles[0] = df_q1
        val_quantiles[1] = df_q2
        val_quantiles[2] = df_q3
        val_quantiles[3] = df_q4
                    
        # Création dictionnaire poiur synthétiser les résultats.

        d= {quantile: ['q_1', 'q_2', 'q_3', 'q_4'],'val_quantiles': val_quantiles}


        # Présentation des résultats dans un dataframe.
        df_results = pd.DataFrame(d)
        
        
        # Création liste pour contenir les dataframes.
        l_df.append(df_results)
        
    
    return l_df
```
